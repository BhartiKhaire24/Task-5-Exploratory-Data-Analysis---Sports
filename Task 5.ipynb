## Data Preparation and Cleaning

```python
matches.duplicated().sum()
```

```python
deliveries.duplicated().sum()
```

```python
deliveries.drop_duplicates(inplace = True)

deliveries.duplicated().sum()
```

```python
# Checking the Null values
matches.isnull().sum()
```

```python
matches.dropna(inplace = True)
```

```python
deliveries.isnull().sum()
```

```python
matches.replace(['Mumbai Indians','Kolkata Knight Riders','Royal Challengers Bangalore','Deccan Chargers','Chennai Super Kings',
                 'Rajasthan Royals','Delhi Daredevils','Gujarat Lions','Kings XI Punjab',
                 'Sunrisers Hyderabad','Rising Pune Supergiants','Kochi Tuskers Kerala','Pune Warriors','Rising Pune Supergiant']
                ,['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW','RPS'],inplace=True)
```

```python
#Replacing the Full names of IPL TEAMS by short names
deliveries.replace(['Mumbai Indians','Kolkata Knight Riders','Royal Challengers Bangalore','Deccan Chargers','Chennai Super Kings',
                 'Rajasthan Royals','Delhi Daredevils','Gujarat Lions','Kings XI Punjab',
                 'Sunrisers Hyderabad','Rising Pune Supergiants','Kochi Tuskers Kerala','Pune Warriors','Rising Pune Supergiant']
                ,['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW','RPS'],inplace=True)
```

```python
# merging seasons column in deliveries dataset which will be helpful in further analysis for each season
deliveries = deliveries.merge(matches["season"], left_on = deliveries.index, right_on=matches.index)
deliveries
```

```python
teams_per_season = matches.groupby('season')['winner'].value_counts()
teams_per_season
```

```python
venue = matches['venue'].value_counts()
plt.title("IPL Venue")
venue.plot(kind = 'barh')
plt.show()
```

#### Wankhede Stadium and Eden Garden has hosted maximum number of IPL matches.

```python
plt.figure(figsize = (7,5))
sns.countplot(x = 'winner', data = matches, order = matches['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Total Victories of IPL Teams")
plt.show()

#team_win_df = matches['winner'].value_counts()
#team_win_df.plot(kind= 'barh')

```

#### Mumbai Indians and chennai Super kings won the maximum no.of IPL.

```python
plt.figure(figsize = (10,5))
best_player = matches.groupby('player_of_match')['season'].sum()
top_10_player = best_player.sort_values(ascending = False).iloc[0:10]
top_10_player.plot(kind = 'barh')
plt.title('Most valuable Player')
plt.show()
```

#### AD Russell, AB de Villiers, Watson, Rashid Khan are the valuable players.
#### Six Indian players have figured in the top ten IPL players list.

```python
# The team that won the most no.of Toss

plt.figure(figsize = (5,7))
sns.countplot(x = 'toss_winner', data = matches, order = matches['toss_winner'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

#### CSK has won the most toss.
